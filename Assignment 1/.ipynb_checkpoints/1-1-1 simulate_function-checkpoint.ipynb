```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import matplotlib.pyplot as plt
import os
torch.manual_seed(1)
```

```python
def train_model(model, inputs, outputs):
    criterion = nn.MSELoss()
    optimizer = optim.Adam(model.parameters(), lr=1e-4)
    max_epochs = 20000
    losses = []

    for epoch in range(max_epochs):
        model.train()
        predictions = model(inputs)
        loss = criterion(predictions, outputs)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        losses.append(loss.item())

        if epoch % 500 == 0 and epoch > 0:
            print(f'epoch {epoch}, loss {loss.item()}')
        if losses[-1] < 0.001:
            print("Change of Loss is Converging and is below threshold value, stopping training")
            break

    return losses
```

```python
class model1(nn.Module):
    def __init__(self):
        super(model1, self).__init__()
        self.linear1 = nn.Linear(1, 8)
        self.linear2 = nn.Linear(8, 16)
        self.linear3 = nn.Linear(16, 16)
        self.linear4 = nn.Linear(16, 16)
        self.linear5 = nn.Linear(16, 16)
        self.linear6 = nn.Linear(16, 16)
        self.linear7 = nn.Linear(16, 8)
        self.linear8 = nn.Linear(8, 1)

    def forward(self, x):
        x = F.leaky_relu(self.linear1(x))
        x = F.leaky_relu(self.linear2(x))
        x = F.leaky_relu(self.linear3(x))
        x = F.leaky_relu(self.linear4(x))
        x = F.leaky_relu(self.linear5(x))
        x = F.leaky_relu(self.linear6(x))
        x = F.leaky_relu(self.linear7(x))
        x = self.linear8(x) 
        return x
```

```python
class model2(nn.Module):
    def __init__(self):
        super(model2, self).__init__()
        self.linear1 = nn.Linear(1, 10)
        self.linear2 = nn.Linear(10, 18)
        self.linear3 = nn.Linear(18, 15)
        self.linear4 = nn.Linear(15, 4)
        self.linear5 = nn.Linear(4, 1)

    def forward(self, x):
        x = F.leaky_relu(self.linear1(x))
        x = F.leaky_relu(self.linear2(x))
        x = F.leaky_relu(self.linear3(x))
        x = F.leaky_relu(self.linear4(x))
        x = self.linear5(x) 
        return x
```

```python
class model3(nn.Module):
    def __init__(self,):
        super(model3, self).__init__()
        self.linear1 = nn.Linear(1, 190)
        self.linear2 = nn.Linear(190, 1)
    
    def forward(self,x):
        x = F.leaky_relu(self.linear1(x))
        x = self.linear2(x) 
        return x
```

```python
# Function to compute sinc function
def sinc_function(x):
    return np.sin(5 * np.pi * x) / (5 * np.pi * x)

x_values = np.linspace(-1, 1, 2000).reshape(-1, 1)

y_values = sinc_function(x_values)

x_tensor = torch.tensor(x_values, dtype=torch.float32)
y_tensor = torch.tensor(y_values, dtype=torch.float32)

plt.plot(x_tensor, y_tensor)
plt.title(r'Sinc-like Function: $\frac{\sin(5\pi x)}{5\pi x}$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```

```python
model_one = model1()
model_one_losses = train_model(model_one, x_tensor, y_tensor)
```

```python
model_two = model2()
model_two_losses = train_model(model_two, x_tensor, y_tensor)
```

```python
model_three = model3()
model_three_losses = train_model(model_three, x_tensor, y_tensor)
```

```python
plt.figure(figsize=(10, 6))
plt.plot(model_one_losses, label='Model 1')
plt.plot(model_two_losses, label='Model 2')
plt.plot(model_three_losses, label='Model 3')
plt.xlabel('Epochs')
plt.ylabel('Loss')
plt.legend()
plt.show()
```

```python
with torch.no_grad():
    model_one.eval()
    model_two.eval()
    model_three.eval()

    predictions_model1 = model_one(x_tensor)
    predictions_model2 = model_two(x_tensor)
    predictions_model3 = model_three(x_tensor)

plt.figure(figsize=(10, 6))
plt.scatter(x_tensor, y_tensor, label='Ground Truth', marker='.', color="y")
plt.plot(x_tensor, predictions_model1.numpy(), label='Model 1', color="r")
plt.plot(x_tensor, predictions_model2.numpy(), label='Model 2', color="b")
plt.plot(x_tensor, predictions_model3.numpy(), label='Model 3', color="g")
plt.xlabel('Input')
plt.ylabel('Output')
plt.legend()
plt.show()
```

```python
def sign_function(x):
    return np.sign(np.sin(5 * np.pi * x))

x_values = np.linspace(-1, 1, 2000).reshape(-1, 1)

y_values = sign_function(x_values)

x_tensor = torch.tensor(x_values, dtype=torch.float32)
y_tensor = torch.tensor(y_values, dtype=torch.float32)

plt.plot(x_tensor, y_tensor)
plt.title(r'Sign of $\sin(5\pi x)$')
plt.xlabel('x')
plt.ylabel('y')
plt.grid(True)

plt.show()
```
