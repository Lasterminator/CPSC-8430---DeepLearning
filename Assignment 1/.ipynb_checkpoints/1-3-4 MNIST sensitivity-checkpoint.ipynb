```python
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import DataLoader
from torchvision import datasets, transforms
from torch.autograd import Variable
import torch.nn.functional as F
import numpy as np
import matplotlib.pyplot as plt
import os
import time
```

```python
torch.manual_seed(1)
device = torch.device("cuda") if torch.cuda.is_available() else torch.device("cpu")
```

```python
def dataset_prep(batch_size):
    transform=transforms.Compose([
            transforms.ToTensor(),
            transforms.Normalize((0.1307,), (0.3081,)),
            ])
    train_data = datasets.MNIST('mnist_data', train=True, download=True, transform=transform)
    test_data = datasets.MNIST('mnist_data', train=False, download=True, transform=transform)

    training_DataLoader = DataLoader(train_data, batch_size=batch_size, shuffle=True)
    testing_DataLoader= DataLoader(test_data, batch_size=batch_size, shuffle=True)
    return training_DataLoader, testing_DataLoader
```

```python
class Model(nn.Module):
    def __init__(self):
        super(Model, self).__init__()
        self.conv1 = nn.Conv2d(1, 32, kernel_size=3, stride=1, padding=1)
        self.pool1 = nn.MaxPool2d(kernel_size=2, stride=2)

        self.dense1 = nn.Linear(32*14*14, 128)
        self.dense2 = nn.Linear(128, 10)

    def forward(self, x):
        x = self.conv1(x)
        x = F.relu(x)
        x = self.pool1(x)
        x = x.view(x.shape[0], -1)
        x = self.dense1(x)
        x = self.dense2(x)
        return F.log_softmax(x, dim=1)

def train(model, optimizer, train_loader):
    model.train()
    loss_func = torch.nn.CrossEntropyLoss()
    correct = 0
    total_samples = 0
    batch_loss = []
    count = 0

    start_time = time.time() 

    for i, batch in enumerate(train_loader):
        count += 1
        images, labels = batch[0].to(device), batch[1].to(device)

        output = model(images)
        loss = loss_func(output, labels)

        optimizer.zero_grad()
        loss.backward()
        optimizer.step()

        _, predicted = torch.max(output.data, 1)
        correct += (predicted == labels).sum().item()
        total_samples += labels.size(0)

        batch_loss.append(loss.item())

    elapsed_time = time.time() - start_time
    print(f"Training completed in {elapsed_time:.2f} seconds")

    train_loss = sum(batch_loss) / count
    acc = 100. * correct / total_samples

    return train_loss, acc


def test(model, test_loader):
    model.eval()
    test_losses = []
    loss_func = torch.nn.CrossEntropyLoss()
    correct = 0
    total_samples = 0

    with torch.no_grad(): 
        for batch in test_loader:
            images, labels = batch[0].to(device), batch[1].to(device)

            output = model(images)
            loss = loss_func(output, labels)
            test_losses.append(loss.item())

            _, predicted = torch.max(output, 1)
            correct += (predicted == labels).sum().item()
            total_samples += labels.size(0)

    test_loss = sum(test_losses) / len(test_loader)
    accuracy = 100. * correct / total_samples

    return test_loss, accuracy

def sensitivity(model):
    total_norm = 0
    count = 0
    
    for param in model.parameters():
        if param.grad is not None:
            grad_norm = torch.linalg.norm(param.grad.to(device)).cpu().item()
            total_norm += grad_norm
            count += 1
    
    return total_norm / count if count > 0 else 0
```

```python
epochs = 16
train_loss_dict = {}
train_acc_dict = {}
test_loss_dict = {}
test_acc_dict = {}
model_state = {}
sensitivity_arr = []

for batch_size in [64, 128, 512, 2048, 4096]:
    train_loss_arr = []
    train_acc_arr = []
    test_loss_arr = []
    test_acc_arr = []
    
    training_loader, testing_loader = dataset_prep(batch_size)
    print(f"Training model with batch size: {batch_size}")
    
    model = Model().to(device)
    optimizer = optim.SGD(model.parameters(), lr=1e-2)

    for epoch in range(epochs):
        train_loss, train_acc = train(model, optimizer, training_loader)
        test_loss, test_acc = test(model, testing_loader)
        
        print(f"Epoch {epoch}: Train Loss = {train_loss:.6f}, Test Loss = {test_loss:.6f}, "
              f"Train Accuracy = {train_acc:.6f}%, Test Accuracy = {test_acc:.6f}%")
        
        train_loss_arr.append(train_loss)
        train_acc_arr.append(train_acc)
        test_loss_arr.append(test_loss)
        test_acc_arr.append(test_acc)

    key_str = f"model_{batch_size}"

    train_loss_dict[key_str] = train_loss_arr
    train_acc_dict[key_str] = train_acc_arr
    test_loss_dict[key_str] = test_loss_arr
    test_acc_dict[key_str] = test_acc_arr
    model_state[key_str] = model

    sensitivity_value = sensitivity(model)
    sensitivity_arr.append(sensitivity_value)
```

```python
all_train_loss_arr = [np.mean(item) for item in train_loss_dict.values()]
all_test_loss_arr = [np.mean(item) for item in test_loss_dict.values()]
```

```python
fig, ax1 = plt.subplots(figsize=(12, 6))

batch_sizes = [64, 128, 512, 2048, 4096]

ax1.plot(batch_sizes, all_train_loss_arr, color="red", label='Train Loss')
ax1.plot(batch_sizes, all_test_loss_arr, color="green", label='Test Loss')
ax1.set_xlabel('Batch Sizes')
ax1.set_ylabel('Cross Entropy Loss', color='black')
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
ax2.plot(batch_sizes, sensitivity_arr, color="blue", label='Sensitivity')
ax2.set_ylabel('Sensitivity', color='black')
ax2.legend(loc='upper right')

plt.title('Model Loss and Sensitivity Comparison')
plt.tight_layout()
plt.show()
```

```python
all_train_acc_arr = [np.mean(item) for item in train_acc_dict.values()]
all_test_acc_arr = [np.mean(item) for item in test_acc_dict.values()]
```

```python
batch_sizes = [64, 128, 512, 2048, 4096]

fig, ax1 = plt.subplots(figsize=(12, 6))

ax1.plot(batch_sizes, all_train_acc_arr, color="red", label='Train Accuracy')
ax1.plot(batch_sizes, all_test_acc_arr, color="green", label='Test Accuracy')
ax1.set_xlabel('Batch Sizes')
ax1.set_ylabel('Accuracy', color='black')
ax1.legend(loc='upper left')

ax2 = ax1.twinx()
ax2.plot(batch_sizes, sensitivity_arr, color="blue", label='Sensitivity')
ax2.set_ylabel('Sensitivity', color='black')
ax2.legend(loc='upper right')

plt.title('Model Accuracy and Sensitivity Comparison')
plt.tight_layout()
plt.show()
```
